```python
df.groupby("grupo_despesa").agg(
        valor_total=("valor_total", "sum"),
        qtd_empenhos=("n_empenho", "count")
    )
```

```python
df = df.drop_duplicates()
```

```python
df = df[(df['valor_total'].sort_values() > 0) & (df['valor_empenhado'].sort_values() > 0) & (df['valor_liquidado'].sort_values() > 0)]
```

```python
df['data'] = pd.to_datetime(df['data'], errors="coerce")

df['ano'] = df['data'].dt.year
```

```python
df['ano'].value_counts()

# indicador de risco: valor empenhado / valor liquidado
# serie temporal por unidade gestora e global + previsão de demanda futura (nome: risco de empenho, liquidação e pagamento)
```

```python
fig = px.histogram(df[df['valor_empenhado'].sort_values() < 50000], x='valor_empenhado', nbins=10, title='Distribuição do Valor Empenhado')
fig.show()
```
